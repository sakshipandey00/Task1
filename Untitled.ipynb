YouTube Streamer Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
import seaborn as sns
```

```python
df = pd.read_csv('youtubers_df.csv',encoding='unicode_escape')
#to avoid encoding error, use 'unicode_escape'
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

Data Cleaning and EDA

```python
#drop links column
df.drop(columns='Links',inplace=True)
```

```python
df.rename(columns={'Suscribers':'Subscribers'},inplace=True)
```

```python
df['Categories'].fillna('Unknown',inplace=True)
```

```python
df.head()
```

```python
df['Username'].duplicated().sum()
```

```python
#removing duplicates in the dataset
df.drop_duplicates(subset=['Username'],inplace=True)
```

```python
df.shape
```

Trend Analysis

Top Subscribed Categories

```python
#top 10 subscribed categories
top_sub=df.groupby('Categories')['Subscribers'].sum().sort_values(ascending=False).reset_index()
top_sub.head(10)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = top_sub.head(5), x='Categories', y='Subscribers')
```

Top Viewed Categories

```python
top_views=df.groupby('Categories')['Visits'].sum().sort_values(ascending=False).reset_index()
top_views.head(10)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= top_views.head(), x='Categories',y='Visits')
```

Top Liked Categories

```python
top_likes=df.groupby('Categories')['Likes'].sum().sort_values(ascending=False).reset_index()
top_likes.head(10)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=top_likes.head(),x='Categories',y='Likes')
```

Top Commented Categories

```python
top_comments=df.groupby('Categories')['Comments'].sum().sort_values(ascending=False).reset_index()
top_comments.head(10)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=top_comments.head(),x='Categories',y='Comments')
```

Most Popular Categories

```python
df['Categories'].value_counts().head(10)
```

```python
df['Categories'].value_counts().head(10).plot(kind='bar',color='r')
plt.title('Most popular categories')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()
```

Is there a corerelation between the number of subscribers and the number of likes or comments?

```python
df.corr=df[['Subscribers','Likes','Comments']].corr()
```

```python
df.corr
```

```python
corr_matrix = df.corr
```

```python
corr_matrix
```

```python
ax= sns.heatmap(df.corr)
plt.title('Correlation between subscribers, likes or comments')
plt.show()
```

Analyze the distribution of streamers audience by country.

```python
country_count=df['Country'].value_counts()
```

```python

plt.figure(figsize=(10,7))
country_count[::-1].plot(kind='barh')
plt.title('Count of streamers')
plt.xlabel('Countries')
plt.ylabel('streamers count')
plt.show()
```
